Data pRE PROCESSING

```python
# SEPEFRATE THE DATA AND LABEL
x = wine.drop('quality', axis= 1)
print(x)
```

```python
y = wine['quality'].apply(lambda y_value: 'Good' if y_value>= 7 else 'Bad')
```

```python
print(y)
```

Training and Test Data

```python
x_train , x_test,  y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
print(y.shape, y_train.shape, y_test.shape)
```

RANDOM FOREST CLASSIFIER

```python
model = RandomForestClassifier()
```

```python
model.fit(x_train, y_train)
```

Model Evaluatioion

```python
# accuracy on test data 
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)
```

```python
print('Accuracy :' , test_data_accuracy)
```

Building a Predictive System

```python
input_data = (1.3,0.05,0.0,1.2,0.005,15.0,21.0,0.9946,3.39,0.47,10.0)

#changing the input data to a numpyarray
input_data_as_numpy_array = np.asarray(input_data)

#reshape the data aswe are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

```
